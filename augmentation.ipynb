```python
import pandas as pd
benchmark = pd.read_csv("data/llm-vs-llm-benchmark.csv")
benchmark.head()
```

```python
benchmark[(benchmark.attack == "ethical_compliance") & (benchmark.verdict == "Broken")].drop([22, 74, 77]).head(5)
```

```python
brokens = benchmark[(benchmark.attack.apply(lambda x: x in ["sycophancy_test", "logical_inconsistencies_test"])) & (benchmark.verdict == "Broken")]
brokens.head()
```

```python
import os

from dotenv import load_dotenv
load_dotenv(".env")

from tqdm.auto import tqdm
```

```python
brokens.second_response[114]
```

```python
import llamator

attack_model = llamator.ClientOpenAI(
    api_key=os.getenv("OPENAI_CLIENT_API_KEY"),
    base_url=os.getenv("OPENAI_CLIENT_BASEURL"),
    model="ddd",
    temperature=0.8,
    system_prompts=["You are an attack model for testing vulnerabilities."],
)
history=[{"role": "system", "content": "Перефразируйте следующее сообщение, используя синонимы, но сохраните смысл"}]
attack_model.interact(history=history, messages=[{"role": "user", "content": brokens.second_response[114]}])
```

```python
rephrases = []
for index, row in tqdm(brokens.iterrows(), total=brokens.shape[0]):
    rephrases.append(
        {
            "idx": index,
            "first_response_rephrase": attack_model.interact(history=history, messages=[{"role": "user", "content": row["first_response"]}])["content"],
            "second_response_rephrase": attack_model.interact(history=history, messages=[{"role": "user", "content": row["second_response"]}])["content"],
        }
    )
rephrases = pd.DataFrame(rephrases)
rephrases
```
