# LLM vs LLM: составление бенчмарка

```python
import pandas as pd
from sklearn.metrics import precision_recall_fscore_support
```

## Ethical Compliance

### Статистики

```python
ethical_df = pd.read_excel("data/llm-vs-llm-generated.xlsx", "ethical_compliance")
ethical_df.head()
```

```python
ethical_df.groupby(["attack_model"]).tested_model.value_counts()
```

```python
pd.concat({
    "A": ethical_df.verdict_A.value_counts(), 
    "B": ethical_df.verdict_B.value_counts(), 
    "ASSESSOR": ethical_df.verdict_assessor.value_counts(), 
    }, axis=1)
```

```python
ethical_df.verdict_A = ethical_df.verdict_A.apply(lambda i: 1 if i == "Broken" else 0)
ethical_df.verdict_B = ethical_df.verdict_B.apply(lambda i: 1 if i == "Broken" else 0)
ethical_df.verdict_assessor = ethical_df.verdict_assessor.apply(lambda i: 1 if i == "Broken" else 0)
ethical_df.head()
```

```python
ethical_df.groupby(["verdict_assessor"]).attack_prompt_score.value_counts()
```

Очистка данных от ситуаций, когда запросы атакующей модели оказались мусором

```python
clean = (ethical_df.verdict_assessor != 0) | (ethical_df.attack_prompt_score > 0) 
pd.concat({
    "A": ethical_df[clean].verdict_A.value_counts(), 
    "B": ethical_df[clean].verdict_B.value_counts(), 
    "ASSESSOR": ethical_df[clean].verdict_assessor.value_counts(), 
    }, axis=1)
```

### Метрики для промтов A и B

```python
print("A:", precision_recall_fscore_support(ethical_df.verdict_assessor, ethical_df.verdict_A, average='binary'))
print("B:", precision_recall_fscore_support(ethical_df.verdict_assessor, ethical_df.verdict_B, average='binary'))
```

```python
print("A:", precision_recall_fscore_support(ethical_df[clean].verdict_assessor, ethical_df[clean].verdict_A, average='binary'))
print("B:", precision_recall_fscore_support(ethical_df[clean].verdict_assessor, ethical_df[clean].verdict_B, average='binary'))
```

```python
ethical_scores = []
for attack_model in ethical_df.attack_model.unique():
    for judge_letter in ["A", "B"]:
        p, r, f1, s = precision_recall_fscore_support(
            ethical_df[clean][ethical_df[clean]["attack_model"] == attack_model]["verdict_assessor"], 
            ethical_df[clean][ethical_df[clean]["attack_model"] == attack_model][f"verdict_{judge_letter}"], 
            average='binary')
        ethical_scores.append({
            "model": attack_model,
            "judge": judge_letter,
            "precision": p,
            "recall": r,
            "f1": f1,
        })
ethical_scores = pd.DataFrame(ethical_scores)
ethical_scores.sort_values("f1", ascending=False)
```

Наиболее высокий результат в основном показывают модели-судьи при использовании **системного промта B**, при этом выигрывает модель **`mistral-large-2407`**.

*Замечание: здесь и далее `nvidia/nemotron-4-340b-instruct` не рассматривается, так как с её помощью было сгененировано всего 5 записей.*

```python
pd.concat([
    ethical_df.groupby(["attack_model"]).attack_prompt_score.value_counts(normalize=False, sort=False),
    ethical_df.groupby(["attack_model"]).attack_prompt_score.value_counts(normalize=True, sort=False),
], axis=1)
```

```python
ethical_df.groupby(["attack_model"]).verdict_assessor.value_counts(normalize=True, sort=False)
```

```python
ethical_df.groupby(["attack_model"]).attack_prompt_score.mean().sort_values(ascending=False)
```

Наиболее лучшим образом атакующий промт генерирует модель **`mistral-large-2407`**.

## Составление чистого бенчмарка 

с полями:
 - attack
 - attack_model
 - tested_model
 - first_attack_prompt
 - first_response
 - second_attack_prompt
 - second_response
 - verdict_assessor

```python
ethical_clean = (ethical_df.verdict_assessor != 0) | (ethical_df.attack_prompt_score > 0) 
logical_clean = (logical_df.verdict_assessor != 0) | (logical_df.attack_prompt_score != -1) 
sycophancy_clean = (sycophancy_df.verdict_assessor != 0) | (sycophancy_df.attack_prompt_score != -1) 

ethical_df["attack"] = "ethical_compliance"
logical_df["attack"] = "logical_inconsistencies_test"
sycophancy_df["attack"] = "sycophancy_test"

benchmark = pd.concat([
    ethical_df[ethical_clean], 
    logical_df[logical_clean], 
    sycophancy_df[sycophancy_clean],
    ]).reset_index(drop=True)

benchmark["verdict"] = benchmark["verdict_assessor"].apply(lambda x: "Broken" if x else "Resilient")

benchmark = benchmark[[
    "attack",
    "attack_model",
    "tested_model",
    "first_attack_prompt",
    "first_response",
    "second_attack_prompt",
    "second_response",
    "verdict",
    ]]
benchmark.head()
```

```python
benchmark.groupby(["attack"]).verdict.value_counts()
```

```python
benchmark.to_csv("data/llm-vs-llm-benchmark.csv", index=False)
```
